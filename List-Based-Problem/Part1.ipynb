```python
# 2.WAP to display max element in a list along with its index
l1 = [1,2,300,4,5,67,90]

max_num = l1[0]
idx = 0
for i in range(len(l1)):
    if l1[i] > max_num:
        max_num = l1[i]
        idx = i
print('list :',l1)
print('Max num :',max_num,", index :",idx)
```

```python
# 2.WAP to display min element in a list along with its index
l1 = [10,2,300,4,5,67,90]

min_num = l1[0]
idx = 0
for i in range(len(l1)):
    if l1[i] < min_num:
        min_num = l1[i]
        idx = i
print('List :',l1)
print('min num :',min_num,", index :",idx)     
```

```python
# 3.WAP to display first two max element along with its difference
l1 = [10,2,300,4,5,67,90]

max_1 = l1[0]
max_2 = l1[1]

for i in range(len(l1)):
    if l1[i] > max_1:
        max_2 = max_1
        max_1 = l1[i]
    elif l1[i] > max_2:
        max_2 = l1[i]
print('first max :',max_1) 
print('second max :',max_2) 
```

```python
# 4.WAPTD missing element from the list excluding the number itself
l1 = [2,12,18,27,38,42,57]
for i in range(len(l1)-1):
    fe = l1[i] + 1
    sec = l1[i+1]
    for i in range(fe,sec):
        print(i,end=' ')
```

```python
# 5.WAP to check given list is palindrome or not 
# reads the same forwards and backwards  ex : 1331 ,MOM

l1 = [1,2,3,4,3,2,1]
j = len(l1) - 1

for i in range(len(l1)):
    if l1[i] != l1[j]:
        print('not palindrome')
        break
    j -= 1
else:
    print("palindrome")
```
