```python
def threeSum(nums):  # keep duplicates
        three_sums = []
        for i in range(0, len(nums)):
            for j in range(i+1, len(nums)):
                for k in range(j+1, len(nums)):
                    if nums[i] + nums[j] + nums[k] == 0:
                        three_sums += [[nums[i], nums[j], nums[k]]]
        return three_sums
threeSum([-1, 0, 1, 2, -1, -4,2])
```

```python
def sum3(num):
    res = []
    for i in range(len(num)):
        left = i + 1
        right = len(num)-1
        while (left < right):
            if num[i]+num[left]+ num[right] == 0:
                res += [[num[i],num[left],num[right]]]
            right -= 1
    return res

sum3([-1, 0, 1, 2, -1, -4,2])
```

```python
def sum3(num):
    num.sort() # 1. Sort the array first
    res = []
    
    for i in range(len(num) - 2):
        # Skip duplicate values for the first element
        if i > 0 and num[i] == num[i - 1]:
            continue
            
        left = i + 1
        right = len(num) - 1
        
        while left < right:
            total = num[i] + num[left] + num[right]
            
            if total == 0:
                res.append([num[i], num[left], num[right]])
                
                # Skip duplicate values for left and right pointers
                while left < right and num[left] == num[left + 1]:
                    left += 1
                while left < right and num[right] == num[right - 1]:
                    right -= 1
                    
                left += 1
                right -= 1
            elif total < 0:
                left += 1 # Sum too small, move right
            else:
                right -= 1 # Sum too big, move left
                
    return res

# Output with corrected code: [[-1, -1, 2], [-1, 0, 1]]
print(sum3([-1, 0, 1, 2, -1, -4]))
```

```python
# 17. Container with Most Water
def max_water(arr):
    res = 0
    for i in range(len(arr)):
        for j in range( i+1, len(arr)):
            height = min(arr[i],arr[j])
            width = j - i
            if (height*width) > res :
                res = height*width
    return res

max_water([2, 1, 8, 6, 4, 6, 5, 5])
```

```python
# 19. Merge sorted arrays

arr1 = [1, 3, 5, 7,9]
arr2 = [2, 4, 6, 8,10]

def merge_sorted_arrays(arr1,arr2):
    n = len(arr1)
    m = len(arr2)
    i = j = 0

    merge = []

    while i < n and j < m :
        if arr1[i] < arr2[j] :
            merge.append(arr1[i])
            i += 1
        else:
            merge.append(arr2[j])
            j += 1
    
    while i < n:
        merge.append(arr1[i])
        i += 1

    while j < m:
        merge.append(arr2[j])
        j += 1

    for i in range(n):
        arr1[i] = merge[i]
    for i in range(m):
        arr2[i] = merge[n+i]
    print("arr1:",arr1)
    print("arr2:",arr2)

merge_sorted_arrays(arr1,arr2)
```

```python
# 20. Move zeros efficiently
arr = [1, 2, 0, 4, 3, 0, 5, 0]

def PushZeroToEnd(arr):
    n = len(arr)
    temp = n * [0]
    j = 0
    for i in arr:
        if i != 0:
            temp[j] = i
            j += 1 
    
    while j < n:
        temp[j] = 0
        j += 1
    return temp

PushZeroToEnd(arr)
```
