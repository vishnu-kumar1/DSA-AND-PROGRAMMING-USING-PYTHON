```python
# 1.Find length of string without using len()

s1 = "VishnuKumar"
cnt = 0
for i in s1:
    cnt += 1
print("len of string ->",cnt)
```

```python
# 2.Count vowels and consonants
s1 = "VishnuKumar"
vowel = 0
const = 0
for i in s1:
    if i in "aeiouAEIOU":
        vowel += 1
    else:
        const += 1
print(f"Vowel :{vowel} ,Consonant :{const}")
```

```python
# 3.Reverse a string
s1 = "VishnuKumar"

res = ""

for i in range(len(s1)-1,-1,-1):
    res += s1[i]
print(s1)
print(res)
```

```python
# 4.Check palindrome string
s1 = "nitrin"

j = len(s1)-1

for i in range(len(s1)//2):
    if s1[i] != s1[j]:
        print('Not a Palindrome String')
        break
    j -= 1
else:
    print('Palindrome Number')
```

```python
# 5.Convert lowercase to uppercase
# (65, 'A', 90, 'Z', 97, 'a', 122, 'z')

s1 = "ABC"
res =""
for i in s1:
    res += chr(ord(i) + 32)

print("Before",s1)
print("After",res)
```

```python
# 6.Convert uppercase to lowercase
# (65, 'A', 90, 'Z', 97, 'a', 122, 'z')
s1 = "abc"
res =""
for i in s1:
    res += chr(ord(i) - 32)

print("Before",s1)
print("After",res)
```
