```python
# 1.Find length of string without using len()

s1 = "VishnuKumar"
cnt = 0
for i in s1:
    cnt += 1
print("len of string ->",cnt)
```

```python
# 2.Count vowels and consonants
s1 = "VishnuKumar"
vowel = 0
const = 0
for i in s1:
    if i in "aeiouAEIOU":
        vowel += 1
    else:
        const += 1
print(f"Vowel :{vowel} ,Consonant :{const}")
```

```python
# 3.Reverse a string
s1 = "VishnuKumar"

res = ""

for i in range(len(s1)-1,-1,-1):
    res += s1[i]
print(s1)
print(res)
```

```python
# 4.Check palindrome string
s1 = "nitrin"

j = len(s1)-1

for i in range(len(s1)//2):
    if s1[i] != s1[j]:
        print('Not a Palindrome String')
        break
    j -= 1
else:
    print('Palindrome Number')
```

```python
# 5.Convert lowercase to uppercase
# (65, 'A', 90, 'Z', 97, 'a', 122, 'z')

s1 = "ABC"
res =""
for i in s1:
    res += chr(ord(i) + 32)

print("Before",s1)
print("After",res)
```

```python
# 6.Convert uppercase to lowercase
# (65, 'A', 90, 'Z', 97, 'a', 122, 'z')
s1 = "abc"
res =""
for i in s1:
    res += chr(ord(i) - 32)

print("Before",s1)
print("After",res)
```

```python
# 7.Count words in sentence
s1 = "The number of words in a sentence is a fundamental string problem."

sentence = s1.split()
count = 0
for i in sentence:
    count += 1
print(count)
    
```

```python
# 8.Remove spaces from string
text = "  Hello   World  "
res = ""
for i in text:
    if i != ' ':
        res += i
res
```

```python
# 9.Find frequency of characters
# Input: "hello"
# Output: {'h': 1, 'e': 1, 'l': 2, 'o': 1}

Input = "hello"
freq = {}
for i in Input:
    if i in freq:
        freq[i] += 1
    else:
        freq[i]=1
freq
 
```

```python
# 10.Find first non-repeating character
input = "hello hell"
res = {}
for i in input:
    if i == ' ':
        continue
    if i in res:
        res[i] += 1
    else:
        res[i] = 1
        
for chr in res:
    if res[chr] == 1:
        print(chr,"this is ")
```

```python
# 11.Find duplicate characters
input = "hello hell"
res = {}
for i in input:
    if i == ' ':
        continue
    if i in res:
        res[i] += 1
    else:
        res[i] = 1
        
for chr in res:
    if res[chr] >= 2:
        print(chr,"this is ")
```
