```python
# 2.Count vowels and consonants
s1 = "VishnuKumar"
vowel = 0
const = 0
for i in s1:
    if i in "aeiouAEIOU":
        vowel += 1
    else:
        const += 1
print(f"Vowel :{vowel} ,Consonant :{const}")
```

```python
# 3.Reverse a string
s1 = "VishnuKumar"

res = ""

for i in range(len(s1)-1,-1,-1):
    res += s1[i]
print(s1)
print(res)
```

```python
# 5.Convert lowercase to uppercase
# (65, 'A', 90, 'Z', 97, 'a', 122, 'z')

s1 = "ABC"
res =""
for i in s1:
    res += chr(ord(i) + 32)

print("Before",s1)
print("After",res)
```

```python
# 9.Find frequency of characters
# Input: "hello"
# Output: {'h': 1, 'e': 1, 'l': 2, 'o': 1}

Input = "hello"
freq = {}
for i in Input:
    if i in freq:
        freq[i] += 1
    else:
        freq[i]=1
freq
 
```

```python
# 10.Find first non-repeating character
input = "hello hell"
res = {}
for i in input:
    if i == ' ':
        continue
    if i in res:
        res[i] += 1
    else:
        res[i] = 1
        
for chr in res:
    if res[chr] == 1:
        print(chr,"this is ")
```

```python
# 11.Find duplicate characters
input = "hello hell"
res = {}
for i in input:
    if i == ' ':
        continue
    if i in res:
        res[i] += 1
    else:
        res[i] = 1
        
for chr in res:
    if res[chr] >= 2:
        print(chr,"this is ")
```

```python
input = "Vishnu Kumar@*&14"
upper = 0
lower = 0
number = 0
spc_chr = 0
for i in input:
    if 65 <= ord(i) <= 90 :
        upper += 1
    elif 97 <= ord(i) <= 122 :
        lower += 1
    elif 48 <= ord(i) <= 57:
        number += 1
    else :
        spc_chr += 1

upper,lower,number,spc_chr
```

```python
# 14.Count digits, alphabets, symbols
input = "Vishnu kumar@!%198"
digit,alpha,symbol = 0,0,0
for i in input:
    if (ord(i)>=ord('A') and ord(i)<=ord('Z')) or (ord(i)>=ord('a') and ord(i)<=ord('z')) :
        alpha += 1
    elif (ord(i)>=ord('0') and ord(i)<=ord('9')):
        digit += 1
    else:
        symbol += 1

alpha,digit,symbol
```

```python
# 15.Find largest word in sentence
text_input = "Your current code has a couple of structural logic errors that prevent it from working ."
list_word = list(text_input.split())

max_length = 0
max_word = None

for word in list_word:
    current_length = len(word)
    if current_length > max_length:
        max_length = current_length
        max_word = word

print(max_word) 
```

```python
# 16.Find smallest word 
text_input = "yours current code couple structural logic errors that prevent from working "
list_word = list(text_input.split())

# Initialize both trackers using the first word in the list
min_length = len(list_word[0])
small_word = list_word[0]

for word in list_word:
    current_length = len(word)
    # Check if the current word is strictly shorter than our minimum found so far
    if current_length < min_length:
        min_length = current_length
        small_word = word

print(small_word)  
```
