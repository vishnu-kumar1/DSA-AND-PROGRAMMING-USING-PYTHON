```python
row = 4
col = 5
val = 1
for i in range(row):
    for j in range(col):
        print(val,end=' ')
        
    print()
    val += 1
    if val > 9:
        val = 1
```

```python
row = 4
col = 4
val = 1
for i in range(row):
    for j in range(col):
        if i == j:
            print(f"*",end=' ')
        else:
            print(' ',end=' ')
        
    print()
    val += 1
    if val > 9:
        val = 1
```
