```python
row = 4
col = 5
val = 1
for i in range(row):
    for j in range(col):
        print(val,end=' ')
        
    print()
    val += 1
    if val > 9:
        val = 1
```

```python
row = 5
col = 5
val = 1
for i in range(row):
    for j in range(col):
        if i >= j:
            print(f"*",end=' ')
        else:
            print(' ',end=' ')
        
    print()
    val += 1
    if val > 9:
        val = 1
```

```python
row = 5
col = 5
val = 1
for i in range(row):
    for j in range(col):
        if i <= j:
            print(f"*",end=' ')
        else:
            print(' ',end=' ')
        
    print()
    val += 1
    if val > 9:
        val = 1
```

```python
val = 1
for i in range(1,5+1):
    for j in range(1,i+1):
        print(val,end=' ')
    print()
    val += 1
```

```python

for i in range(1,5+1):
    val = 2
    for j in range(1,i+1):
        print(val,end=' ')
        val += 2
    print()
    if val > 9:
        val = 2
```
