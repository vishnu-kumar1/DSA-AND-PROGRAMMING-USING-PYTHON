```python
# 1. Factorial of given number
def fun(n):
    if n<=1:
        return 1
    else:
        return n*fun(n-1)
    
fun(4)
```

```python
# 2.Print numbers from 1 to N

def print_1_to_n(num):
    if num == 0 : # Base condition : When to stop 
        return 
    print_1_to_n(num-1) # Recursive call 
    print(num) # Print the  number
    
print_1_to_n(5)
```

```python
# 2. Print number from n to 1 

def print_n_to_1(num):

    if num < 1:  # Base condition : When to stop 
        return 1
    print(num)
    print_n_to_1(num -1) # Recursive call 

print_n_to_1(5) # Argument must be positive
```

```python
# 3. Find Power of number

def power_of_num(base , exponent):
    if exponent == 0 : # Base case : when expo is 0 it will return 1 
        return 1
    return base * power_of_num(base,exponent - 1) # Recursive call 

power_of_num(2 ,3)
```

```python
# 4. Sum of first n natural number

def sum_of_n(num):
    if num == 1:  # Base case : When argument is 1 it return 1
        return 1
    
    return num + sum_of_n(num -1)

sum_of_n(5) # 1+2+3+4+5
```

```python
# 5. Sum of digit of number

def sum_of_digit(num):

    if len(str(num)) == 1: #Base condition : when len is 1 return itself
        return num
    
    return num % 10 + sum_of_digit(num//10)

sum_of_digit(123)
```

```python
# 6. reverse a string recursively

def string_rev(word):

    if len(word) == 1 : # Base case : when len of string is 1 return itself 
        return word
    
    return word[-1] + string_rev(word[0:len(word)-1])  #Recursive call 

string_rev('vishnu')
```

```python

# 7. Reverse a number 
def rev_num(num):

    if len(str(num)) == 1:
        return num 
    
    return num % 10 * 10 + rev_num(num // 10)

rev_num(12)
```

```python
n = 1234
rev = 0

while n > 0 :
    ls = n%10 
    rev = (rev * 10) + ls
    n = n // 10

rev
```
