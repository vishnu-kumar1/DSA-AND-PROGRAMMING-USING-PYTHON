```python
# 1. Factorial of given number
def fun(n):
    if n<=1:
        return 1
    else:
        return n*fun(n-1)
    
fun(4)
```

```python
# 2.Print numbers from 1 to N

def print_1_to_n(num):
    if num == 0 :
        return 
    print_1_to_n(num-1)
    print(num)
    
print_1_to_n(2)
```

```python
def print_1_to_n_recursive(n):
    # 1. Base Condition: When to stop
    if n == 0:
        return
    
    # 2. Recursive Call: Reduce the problem size
    print_1_to_n_recursive(n - 1)
    
    # 3. Work: Print the number
    print(n)

# Example usage:
print_1_to_n_recursive(5)
```
