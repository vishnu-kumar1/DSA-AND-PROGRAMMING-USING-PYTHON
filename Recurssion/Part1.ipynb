```python
# 1. Factorial of given number
def fun(n):
    if n<=1:
        return 1
    else:
        return n*fun(n-1)
    
fun(4)
```

```python
def reverse_string(s):
    if len(s) <= 1:        # Base Case
        return s
    return reverse_string(s[1:]) + s[0]  # Recursive Case
```
